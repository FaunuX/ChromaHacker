```python
import requests
import json

def get_image_url():
  response = requests.get("https://commons.wikimedia.org/w/api.php?action=query&generator=random&grnnamespace=6&prop=imageinfo&iiprop=url&format=json").content
  key = list(json.loads(response.decode('utf-8'))['query']['pages'])[0]
  return json.loads(response.decode('utf-8'))['query']['pages'][key]['imageinfo'][0]['url']
```

```python
import random

colorscheme_list = []

def hex_value(val):
  if len(val) == 1:
    return '0' + val
  else:
    return val

def convert_colors(color_tuple):
  return '#' + hex_value(hex(color_tuple[0])[2:]) + hex_value(hex(color_tuple[1])[2:]) + hex_value(hex(color_tuple[2])[2:])

def gen_colorscheme():
  color1 = (random.randint(0, 63), random.randint(0, 63), random.randint(0, 63))
  color2 = (random.randint(64, 127), random.randint(64, 127), random.randint(64, 127))
  color3 = (random.randint(128, 191), random.randint(128, 191), random.randint(128, 191))
  color4 = (random.randint(192, 255), random.randint(192, 255), random.randint(192, 255))
  colorscheme_list.append([color1, color2, color3, color4])
  rv =  (convert_colors(color1), convert_colors(color2), convert_colors(color3), convert_colors(color4))
  print(rv)
  return rv
```

```python
import numpy as np from PIL import Image
def scale(image, size=(32, 32)):
    # Convert the input image to a PIL Image object
    pil_image = Image.fromarray(np.rint(image).astype(np.uint8))

    # Resize the image using the specified size
    scaled_image = pil_image.resize(size, Image.LANCZOS)

    # Convert the scaled PIL Image back to a NumPy array
    scaled_array = np.array(scaled_image)

    return scaled_array
```

```python
import requests
import numpy as np
from PIL import Image
from io import BytesIO
from chromahacker.palettize import palettize

image_arrays = []

for i in image_urls:
  try:
    image_arrays.append(
        scale(
            palettize(i, *gen_colorscheme(), headers={
                'User-Agent': 'Chromahacker'
            }),
            (64, 64)
        )
    )
  except:
    colorscheme_list.pop()

image_arrays = np.array(image_arrays)
```
